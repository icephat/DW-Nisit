```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('asso.csv', names = ['kuSubjectId'], sep = ',')
df.head()
```

```python
df.shape
```

```python
data = list(df["kuSubjectId"].apply(lambda x:x.split("/") ))
data
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df = pd.DataFrame(a_data,columns=a.columns_)
df = df.replace(False,0)
df
```

```python
df = apriori(df, min_support = 0.1, use_colnames = True, verbose = 1)
df
```

```python
#Let's view our interpretation values using the Associan rule function.
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.8)
df_ar
```
